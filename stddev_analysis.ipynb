```python
import numpy as np


def readCSVForSD(filename) :
    data = np.genfromtxt("results/" + "raw/" + filename, names=True, delimiter=',', dtype=None, encoding='utf-8')
    column_names = data.dtype.names
    sizes = data['size']
    dataset = data['dataset']
    stddevs = data['stddev']
    print("==" + filename + "==")
    print("Dataset | Size | Standard Deviation")
    print("-------------------------")

    for d, s, sd in zip(dataset, sizes, stddevs):
        print(f"{d} | {s} | {sd}")

    print("-------------------------")

def aggregateSizeForSD(filenames):
    alreadyAccessed = [];
    sizeMap = dict()
    for i in filenames:
        data = np.genfromtxt("results/" + "raw/" + i, names=True, delimiter=',', dtype=None, encoding='utf-8')
        column_names = data.dtype.names
        datasets = data['dataset']
        sizes = data['size']
        stddevs = data['stddev']
        for d, s, sd in zip(datasets, sizes, stddevs):
            if d in alreadyAccessed: continue
            s = int(s)
            if (sd == "NO_OUTPUT"): continue
            sd = float(sd)
            if s not in sizeMap:
                sizeMap[s] = []    
            sizeMap[s].append(sd)  
    return sizeMap

def printAggregateSD(sizeDict):
    for size, values in sorted(sizeDict.items()):
        avg = np.mean(values)
        print(f"Size {size}: Average StdDev = {avg}")
```

```python
np.set_printoptions(suppress=True)
# readCSVForSD("DT_CES_SA_20251123_050616_9079e4.csv")
# readCSVForSD("DT_CES_SA_20251123_105925_0e2e35.csv")
#readCSVForSD("DT_CES_SA_20251123_150032_a59c33.csv")
readCSVForSD("DT_CES_SA_20251123_150110_0057aa.csv")

sizeDict = aggregateSizeForSD(["DT_CES_SA_20251123_150110_0057aa.csv"])
printAggregateSD(sizeDict)
```
