```python
import numpy as np
import os
def parseFiles():

    filepaths = []
    directory = "results/raw"
    keyword = "DT_CES_SA"
   
    for filename in os.listdir(directory):
       
        if filename.startswith(keyword) and filename.endswith(".csv"):
            filepath = os.path.join(directory, filename)
            filepaths.append(filepath)
    return filepaths

def readCSVForSD(filename) :
    data = np.genfromtxt(f"results/raw/" + filename, names=True, delimiter=',', dtype=None, encoding='utf-8')
    column_names = data.dtype.names
    sizes = data['size']
    dataset = data['dataset']
    stddevs = data['stddev']
    print("==" + filename + "==")
    print("Dataset | Size | Standard Deviation")
    print("-------------------------")

    for d, s, sd in zip(dataset, sizes, stddevs):
        if (sd == "NO_OUTPUT"): continue
        print(f"{d} | {s} | {sd}")

    print("-------------------------")

def aggregateSizeForSD(filenames):
    alreadyAccessed = [];
    sizeMap = dict()
    for i in filenames:
        data = np.genfromtxt(i, names=True, delimiter=',', dtype=None, encoding='utf-8', ndmin=1)
        column_names = data.dtype.names
        datasets = data['dataset']
        sizes = data['size']
        stddevs = data['stddev']
        for d, s, sd in zip(datasets, sizes, stddevs):
            
            if d in alreadyAccessed: continue
            alreadyAccessed.append(d)

            s = int(s)
            if (sd == "NO_OUTPUT"): continue
            if (sd == 0): continue
            sd = float(sd)
            if s not in sizeMap:
                sizeMap[s] = []    
            sizeMap[s].append(sd)  
    return sizeMap

def printAggregateSD(sizeDict):
    print(f"no of points | no of datasets | avg sd")
    for size in sorted(sizeDict.keys()):
        values = sizeDict[size]
  
        avg = np.mean(values)
        print(avg)
        print(f"{size:5d} | {len(values)} | {avg:10.6f}")
```

```python
np.set_printoptions(suppress=True)
# Note: all csvs were picked such that they contained the runs that had multiple tests (no 0 sds)
filenames = parseFiles()
print(len(filenames))
readCSVForSD("DT_CES_SA_20251123_150110_0057aa.csv")
readCSVForSD("DT_CES_SA_20251123_143856_9a6893.csv")
sizeDict = aggregateSizeForSD(parseFiles())
#printAggregateSD(sizeDict)
```
